# CLEANING THE DATASETS

## Traffic Count

```python
def clean_traffic_count(df,list_columns):
    for i in list_columns:
        if type(df[i][0]) == str:
            df[i] = pd.to_numeric(df[i].str.strip()).fillna(0)
        else: 
            df[i] = df[i].fillna(0)
```

```python
cleaned_traffic = traffic_counts.drop(['id','file_no'],axis=1)
```

```python
cleaned_traffic.dtypes
# all_count and northbound_count are objects when they should be floats; we need to convert them
```

```python
traffic_count_cols = ['all_count','northbound_count','southbound_count','eastbound_count','westbound_count']
clean_traffic_count(cleaned_traffic,traffic_count_cols)
```

```python
cleaned_traffic['count_date'] = cleaned_traffic['count_date'].apply(pd.to_datetime)
#Converts count_date to datetime
```

```python
cleaned_traffic['day'] = cleaned_traffic['count_date'].dt.weekday_name
cleaned_traffic
#Adds weekday based on date
```

```python
total_cars = cleaned_traffic.groupby(['day']).mean()
#checks how many cars are driving on that date
total_cars = total_cars.drop(['all_count','northbound_count','southbound_count','eastbound_count','westbound_count'],axis = 1)
total_cars
```

```python
total_cars.plot(kind='bar')
plt.ylabel('Average Cars')
plt.xlabel('Date')
plt.title('Amount of cars a certain day')
plt.show()
#Shows how many cars are driven during a certain day 
```

```python
x = np.log(total_cars['total_count'].values)
total_cars['logged_cars'] = x
logged_cars = total_cars.drop(['total_count'],axis = 1)
```

```python
logged_cars.plot(kind='bar')
plt.ylabel('Total Cars')
plt.xlabel('Date')
plt.title('Amount of cars a certain day')
plt.show()
#Shows how many cars are driven during a certain day 
```

```python
charges = cleaned_collisions.groupby(cleaned_collisions['hit_run_lvl']).sum()
charges
```

```python
cleaned_collisions['hit_run_lvl'].value_counts()
police_beats['beat'].unique()
#Checking what are all the beats
```

```python
lists = police_beats['beat'].unique().tolist()
bool_array = cleaned_collisions['police_beat'].apply(lambda x: x in lists)
cleaned_beats = cleaned_collisions[bool_array]
cleaned_beats
#Checking invalid beats
```

```python
cleaned_beats['day'] = cleaned_beats['date_time'].dt.weekday_name
total_collisions = cleaned_beats.groupby(['day']).sum()
#checks how many cars are driving on that date
total_collisions = total_collisions.drop(['police_beat','street_no'],axis = 1)
total_collisions['injury'] = total_collisions['injured'].apply(lambda x: x/total_collisions['injured'].sum())
total_collisions['kill'] = total_collisions['killed'].apply(lambda x: x/total_collisions['killed'].sum())
total_collisions
#Finding percent of people killed on what day
```

```python
proportion_collisions = total_collisions.drop(['injured', 'killed'],axis = 1)
proportion_collisions
```
